```python
from gurobipy import *
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#reading the data
customers = pd.read_csv('data/customer_locations.csv', index_col=0)
warehouses = pd.read_csv('data/warehouse_locations.csv', index_col=0)
supply = pd.read_csv('data/supply.csv', index_col=0)
```

```python
#caculating distances
def euclidean_distance(x1, y1, x2, y2):
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)
```

```python
#distance matrix:
distances = {}
for i in warehouses.index:
    for j in customers.index:
        distances[i,j] = euclidean_distance(
            warehouses.iloc[i,0], warehouses.iloc[i,1],
            customers.iloc[j,0], customers.iloc[j,1]
        )
```

```python
#make opt model
m = Model('delivery_optimization')
```

```python
#create variables
x = {}
for i in warehouses.index:
    for j in customers.index:
        x[i,j] = m.addVar(vtype=GRB.BINARY, name=f'x_{i}_{j}')
```

```python
#set objective
m.setObjective(quicksum(distances[i,j] * x[i,j] for i in warehouses.index for j in customers.index), GRB.MINIMIZE)
```

```python
#constraints
k = 6

#amount of deliveries constraints
m.addConstr(quicksum(x[i,j] for i in warehouses.index for j in customers.index) == k, "driver_count")

#customers receive at most one delivery
for j in customers.index:
    m.addConstr(quicksum(x[i,j] for i in warehouses.index) <= 1, f"customer_{j}")

#warehouse supply constraint
for i in warehouses.index:
    m.addConstr(quicksum(x[i,j] for j in customers.index) <= supply.iloc[i,0], f"warehouse_{i}")
```

```python
#optimize
m.optimize()
```

```python
#results
print("\nDelivery Schedule:")
print("Warehouse -> Customer: Distance")
total_distance = 0
for i in warehouses.index:
    for j in customers.index:
        if x[i,j].X > 0.5:  # If assignment is made
            print(f"Warehouse {i} -> Customer {j}: {distances[i,j]:.2f}")
            total_distance += distances[i,j]

print(f"\nTotal Distance: {total_distance:.2f}")
print(f"Number of customers served: {sum(x[i,j].X > 0.5 for i in warehouses.index for j in customers.index)}")
```

```python
#distance matrix
w2c_dist = {}
for i in warehouses.index:
    for j in customers.index:
        w2c_dist[i,j] = euclidean_distance(
            warehouses.iloc[i,0], warehouses.iloc[i,1],
            customers.iloc[j,0], customers.iloc[j,1]
        )

#customer to customer distance
c2c_dist = {}
for i in customers.index:
    for j in customers.index:
        if i != j:
            c2c_dist[i,j] = euclidean_distance(
                customers.iloc[i,0], customers.iloc[i,1],
                customers.iloc[j,0], customers.iloc[j,1]
            )
```

```python
#create model
m = Model('multi_delivery_optimization')
```

```python
#decision Variables
x = {}
for i in warehouses.index:
    for j in customers.index:
        x[i,j] = m.addVar(vtype=GRB.BINARY, name=f'x_{i}_{j}')

y = {}
for i in customers.index:
    for j in customers.index:
        if i != j:
            y[i,j] = m.addVar(vtype=GRB.BINARY, name=f'y_{i}_{j}')

u = {}
for i in customers.index:
    u[i] = m.addVar(lb=1, ub=5, vtype=GRB.INTEGER, name=f'u_{i}')

z = {}
for i in customers.index:
    z[i] = m.addVar(vtype=GRB.BINARY, name=f'z_{i}')
```

```python
#objective -> minimize distances
obj = quicksum(w2c_dist[i,j] * x[i,j] for i in warehouses.index for j in customers.index)  # Warehouse to first customer
obj += quicksum(c2c_dist[i,j] * y[i,j] for i in customers.index for j in customers.index if i != j)  # Between customers
m.setObjective(obj, GRB.MINIMIZE)
```

```python
#constraints
for j in customers.index:
    m.addConstr(
        quicksum(x[i,j] for i in warehouses.index) + 
        quicksum(y[i,j] for i in customers.index if i != j) == 1
    )
```

```python
#flow of conservation
for i in customers.index:
    m.addConstr(
        quicksum(y[i,j] for j in customers.index if i != j) + z[i] == 
        quicksum(x[w,i] for w in warehouses.index) + 
        quicksum(y[j,i] for j in customers.index if j != i)
    )

#warehouse capacity constraints
for i in warehouses.index:
    m.addConstr(quicksum(x[i,j] for j in customers.index) <= supply.iloc[i,0])

#subtour
M = len(customers)
for i in customers.index:
    for j in customers.index:
        if i != j:
            m.addConstr(u[i] - u[j] + M * y[i,j] <= M - 1)

#route length
for i in customers.index:
    m.addConstr(u[i] <= 5)
```

```python
#optimize model
m.optimize()
```

```python
#printing results
print("\nOPTIMAL DELIVERY SCHEDULE:")
print("==========================")

route_count = 0
customers_served = set()
route_lengths = []

for i in warehouses.index:
    for j in customers.index:
        if x[i,j].X > 0.5:
            route_count += 1
            current_route_length = 1
            print(f"\nRoute {route_count} (Starting from Warehouse {i}):")
            print(f"1. Warehouse {i} -> Customer {j}")
            total_distance += w2c_dist[i,j]
            customers_served.add(j)
            
            current = j
            route_finished = False
            stop_number = 2
            
            while not route_finished:
                route_finished = True
                for next_cust in customers.index:
                    if next_cust != current and y[current,next_cust].X > 0.5:
                        print(f"{stop_number}. Customer {current} -> Customer {next_cust}")
                        total_distance += c2c_dist[current,next_cust]
                        current = next_cust
                        customers_served.add(next_cust)
                        route_finished = False
                        current_route_length += 1
                        stop_number += 1
                        break
                if z[current].X > 0.5:
                    print(f"   End at Customer {current}")
                    route_finished = True
            
            route_lengths.append(current_route_length)

print("\nSUMMARY STATISTICS:")
print("===================")
print(f"Total Distance: {total_distance:.2f} units")
print(f"Number of Routes: {route_count}")
print(f"Total Customers Served: {len(customers_served)}")
print(f"Average Stops per Route: {sum(route_lengths)/len(route_lengths):.1f}")
print(f"Route Lengths: {route_lengths}")
```

```python
#distance matrix
w2c_dist = {}
for i in warehouses.index:
    for j in customers.index:
        w2c_dist[i,j] = euclidean_distance(
            warehouses.iloc[i,0], warehouses.iloc[i,1],
            customers.iloc[j,0], customers.iloc[j,1]
        )

#customer to customer distance
c2c_dist = {}
for i in customers.index:
    for j in customers.index:
        if i != j:
            c2c_dist[i,j] = euclidean_distance(
                customers.iloc[i,0], customers.iloc[i,1],
                customers.iloc[j,0], customers.iloc[j,1]
            )

#create model
m = Model('multi_delivery_optimization')

#decision Variables (same as before)
x = {}
for i in warehouses.index:
    for j in customers.index:
        x[i,j] = m.addVar(vtype=GRB.BINARY, name=f'x_{i}_{j}')

y = {}
for i in customers.index:
    for j in customers.index:
        if i != j:
            y[i,j] = m.addVar(vtype=GRB.BINARY, name=f'y_{i}_{j}')

u = {}
for i in customers.index:
    u[i] = m.addVar(lb=1, ub=5, vtype=GRB.INTEGER, name=f'u_{i}')

z = {}
for i in customers.index:
    z[i] = m.addVar(vtype=GRB.BINARY, name=f'z_{i}')

# Add new variable to track which warehouses are used as starting points
w = {}
for i in warehouses.index:
    w[i] = m.addVar(vtype=GRB.BINARY, name=f'w_{i}')

#objective -> minimize distances (same as before)
obj = quicksum(w2c_dist[i,j] * x[i,j] for i in warehouses.index for j in customers.index)
obj += quicksum(c2c_dist[i,j] * y[i,j] for i in customers.index for j in customers.index if i != j)
m.setObjective(obj, GRB.MINIMIZE)

#constraints (same as before)
for j in customers.index:
    m.addConstr(
        quicksum(x[i,j] for i in warehouses.index) + 
        quicksum(y[i,j] for i in customers.index if i != j) == 1
    )

#flow conservation (same as before)
for i in customers.index:
    m.addConstr(
        quicksum(y[i,j] for j in customers.index if i != j) + z[i] == 
        quicksum(x[w,i] for w in warehouses.index) + 
        quicksum(y[j,i] for j in customers.index if j != i)
    )

#warehouse capacity constraints (same as before)
for i in warehouses.index:
    m.addConstr(quicksum(x[i,j] for j in customers.index) <= supply.iloc[i,0])

#subtour (same as before)
M = len(customers)
for i in customers.index:
    for j in customers.index:
        if i != j:
            m.addConstr(u[i] - u[j] + M * y[i,j] <= M - 1)

#route length (same as before)
for i in customers.index:
    m.addConstr(u[i] <= 5)

# New constraints to handle k drivers
k_drivers = 5  # Set this to your desired number of drivers

# Link warehouse usage to route starts
for i in warehouses.index:
    m.addConstr(quicksum(x[i,j] for j in customers.index) <= M * w[i])
    
# Limit total number of drivers/routes
m.addConstr(quicksum(w[i] for i in warehouses.index) <= k_drivers)

#optimize model
m.optimize()
```

```python
#printing results
print("\nOPTIMAL DELIVERY SCHEDULE:")
print("==========================")

route_count = 0
customers_served = set()
route_lengths = []

for i in warehouses.index:
    for j in customers.index:
        if x[i,j].X > 0.5:
            route_count += 1
            current_route_length = 1
            print(f"\nRoute {route_count} (Starting from Warehouse {i}):")
            print(f"1. Warehouse {i} -> Customer {j}")
            total_distance += w2c_dist[i,j]
            customers_served.add(j)
            
            current = j
            route_finished = False
            stop_number = 2
            
            while not route_finished:
                route_finished = True
                for next_cust in customers.index:
                    if next_cust != current and y[current,next_cust].X > 0.5:
                        print(f"{stop_number}. Customer {current} -> Customer {next_cust}")
                        total_distance += c2c_dist[current,next_cust]
                        current = next_cust
                        customers_served.add(next_cust)
                        route_finished = False
                        current_route_length += 1
                        stop_number += 1
                        break
                if z[current].X > 0.5:
                    print(f"   End at Customer {current}")
                    route_finished = True
            
            route_lengths.append(current_route_length)

print("\nSUMMARY STATISTICS:")
print("===================")
print(f"Total Distance: {total_distance:.2f} units")
print(f"Number of Routes: {route_count}")
print(f"Total Customers Served: {len(customers_served)}")
print(f"Average Stops per Route: {sum(route_lengths)/len(route_lengths):.1f}")
print(f"Route Lengths: {route_lengths}")
```

```python
k_values = range(4, 11)
total_distances = []

#optimization for each value
for k in k_values:
    m = Model('multi_delivery_optimization')
    m.setParam('OutputFlag', 0) #not print gurobi
    
    #decision variables
    x = {}
    for i in warehouses.index:
        for j in customers.index:
            x[i,j] = m.addVar(vtype=GRB.BINARY, name=f'x_{i}_{j}')

    y = {}
    for i in customers.index:
        for j in customers.index:
            if i != j:
                y[i,j] = m.addVar(vtype=GRB.BINARY, name=f'y_{i}_{j}')

    u = {}
    for i in customers.index:
        u[i] = m.addVar(lb=1, ub=5, vtype=GRB.INTEGER, name=f'u_{i}')

    z = {}
    for i in customers.index:
        z[i] = m.addVar(vtype=GRB.BINARY, name=f'z_{i}')

    #objective
    obj = quicksum(w2c_dist[i,j] * x[i,j] for i in warehouses.index for j in customers.index)
    obj += quicksum(c2c_dist[i,j] * y[i,j] for i in customers.index for j in customers.index if i != j)
    m.setObjective(obj, GRB.MINIMIZE)

    #constraints
    m.addConstr(quicksum(x[i,j] for i in warehouses.index for j in customers.index) == k)

    for j in customers.index:
        m.addConstr(
            quicksum(x[i,j] for i in warehouses.index) + 
            quicksum(y[i,j] for i in customers.index if i != j) == 1
        )

    for i in customers.index:
        m.addConstr(
            quicksum(y[i,j] for j in customers.index if i != j) + z[i] == 
            quicksum(x[w,i] for w in warehouses.index) + 
            quicksum(y[j,i] for j in customers.index if j != i)
        )

    for i in warehouses.index:
        m.addConstr(quicksum(x[i,j] for j in customers.index) <= supply.iloc[i,0])

    M = len(customers)
    for i in customers.index:
        for j in customers.index:
            if i != j:
                m.addConstr(u[i] - u[j] + M * y[i,j] <= M - 1)

    for i in customers.index:
        m.addConstr(u[i] <= 5)

    #optimize
    m.optimize()

    #calculate total distance for k
    if m.status == GRB.OPTIMAL:
        total_distance = 0
        #warehouse to first customer distance
        for i in warehouses.index:
            for j in customers.index:
                if x[i,j].X > 0.5:
                    total_distance += w2c_dist[i,j]
        
        #customer to customer distance
        for i in customers.index:
            for j in customers.index:
                if i != j and y[i,j].X > 0.5:
                    total_distance += c2c_dist[i,j]
        
        total_distances.append(total_distance)
    else:
        total_distances.append(None)
```

```python
#plot
plt.figure(figsize=(10, 6))
plt.plot(k_values, total_distances, 'bo-')

#labels
plt.xlabel('Number of Drivers')
plt.ylabel('Total Distance')
plt.title('Total Distance vs Number of Drivers')

#grid
plt.grid(True)

#show plot
plt.show()
```
